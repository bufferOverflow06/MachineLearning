```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
```

### MultiLinearRegression

```python
df = pd.read_csv('multiple_linear_regression_dataset.csv')
print(df.head())
print(df.isnull().sum())
print(df.shape)
```

```python
df1 = df
df1.duplicated(subset='income')
df1.drop_duplicates(subset='income',keep='last',inplace=False)
print(df1.shape)
```

```python
X_age = df.iloc[:,0]
y = df.iloc[:,-1]
print(X_age.shape)
print(y.shape)
plt.scatter(X_age,y,color='g',marker='+')
plt.xlabel('age')
plt.ylabel('income')
plt.show()

X_age = np.asanyarray(X_age).reshape(-1,1)
lin_reg = LinearRegression()
reg_age = lin_reg.fit(X_age,y)
pred_age = reg_age.predict(X_age)
#print(pred)


plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.scatter(X_age,y,color='g',marker='+')
plt.plot(X_age,pred_age,color='red')
```
