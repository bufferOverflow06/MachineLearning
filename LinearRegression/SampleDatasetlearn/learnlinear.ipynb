```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv('current.csv')
df.head()
```

```python
%matplotlib inline

plt.xlabel('area(sq. ft)')
plt.ylabel('price($)')
plt.scatter(df.area,df.price,color='r',marker='+')
```

```python
lin_reg = linear_model.LinearRegression()
lin_reg.fit(df[['area']],df.price)
```

```python
lin_reg.predict(np.array([3000]).reshape(-1,1)) # first argument for both fit and predict should be 2 dimensional
```

```python
print('coef: ',lin_reg.coef_)
print('intercept: ',lin_reg.intercept_)
```

```python
plt.xlabel('area(sq. ft)')
plt.ylabel('price($)')
plt.scatter(df.area,df.price,color='r',marker='+')
plt.plot(df.area,lin_reg.predict(df[['area']]),color='b')
```

#### Predicting prices to csv file

```python
dfp = pd.read_csv('predictions.csv')
dfp.head(3)
pred = lin_reg.predict(dfp)
```

```python
dfp['predicted_prices'] = pred


dfp.to_csv('predictions.csv',index=False)
dfp
```

### Canada_per_capita_income perdiction

```python
dataframe = pd.read_csv('canada_per_capita.csv')
dataframe.head(4)
```

```python
dataframe.isnull().sum()
```

```python
plt.xlabel('year')
plt.ylabel('percapita')
y_axis = dataframe.per_capita_income
x_axis = dataframe.year
plt.scatter(x_axis,y_axis,color='g',marker='+')
```

```python
model = linear_model.LinearRegression()
model.fit(dataframe[['year']],dataframe.per_capita_income)
model.predict([[2020]])
```

```python
plt.xlabel('year')
plt.ylabel('percapita')
x_axis = dataframe.year
y_axis = dataframe.per_capita_income
predicted_values = model.predict(dataframe[['year']])
plt.scatter(x_axis,y_axis,color='g',marker='+')
plt.plot(dataframe.year,predicted_values,color = 'red')
```
